Write a function in recursion to detect sequence "C A T C A T" in above linked list format of DNA by det(ll) function call

```python
# Answer:
empty = []

def is_LL(s):
    return s == empty or (len(s) == 2 and is_LL(s[1]))

def LL(first, rest):
    assert is_LL(rest), "rest must be a singly linked list"
    return [first, rest]


gene = 'CATCAT'

# helper function
def detect_gene(dna, gene, l):

    if l==len(gene):
        return True

    tmp = dna
    if tmp == empty:
        return False

    if tmp[0] == gene[l]:
        return detect_gene(tmp[1], gene, l+1)
    else:
        return detect_gene(tmp[1], gene, 0)


def det(dna):
    return detect_gene(dna, gene, 0)


dna = LL('C', LL('A', LL('T', empty)))
dna = LL('C', LL('A', LL('T', LL('G', dna))))
print(dna)
# ['C', ['A', ['T', ['G', ['C', ['A', ['T', []]]]]]]]

print(det(dna))

end = LL('T', LL('C', LL('A', LL('T', LL('G', empty)))))
dna = LL('G', LL('T', LL('A', LL('C', LL('A', end)))))
print(dna)
#['G', ['T', ['A', ['C', ['A', ['T', ['C', ['A', ['T', ['G', []]]]]]]]]]]

print(det(dna))

print(det(end))
```

**Q2**:	Assuming that there exists a singly linked list with int type nodes, such as 
ls = Head->1->2->3->4->5->6->None,  create a function/method to separate it to two singly linked lists and the first half should be in reverse sequence, but the second half keeps original chain. For instance, two new singly linked lists will be 
ls1 = Head->3->2->1->None, and ls2 = Head->4->5->6->None  from function call sep(ls).

Notice that if the number of nodes in original list is odd, delete the middle one and generate two new lists following above rules, such as 
ls = Head->1->2->3->4->5->6->7->None, after calling sep(ls), you will get 
ls1 = Head->3->2->1->None, and ls2 = Head->5->6->7->None, where value 4 of the middle node is deleted


```python
# Answer:
class Node:
    def __init__(self, data):
        self.data = data  # Assign data
        self.next = None  # Initialize next as null


# Linked List class contains a Node object
class LinkedList:

    def __init__(self):
        self.head = None

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def append(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return

        last = self.head
        while (last.next):
            last = last.next

        last.next = new_node

    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data, end=' ')
            temp = temp.next
        print("")


def reverse_LL(nd):
    prev = None
    current = nd
    while current is not None:
        next = current.next
        current.next = prev
        prev = current
        current = next
    return prev


def sep(ll):
    c = 0
    current = ll.head
    while current is not None:
        current = current.next
        c += 1

    if c < 2:
        return None, None

    if c % 2 == 0: #even
        current = ll.head
        for i in range(0, int(c/2)-1):
            current = current.next

        second = LinkedList()
        second.head = current.next
        current.next = None
        first = LinkedList()
        first.head = reverse_LL(ll.head)

    else:
        current = ll.head
        for i in range(0, int(c / 2) - 1):
            current = current.next

        second = LinkedList()
        second.head = current.next.next
        current.next.next = None
        current.next = None
        first = LinkedList()
        first.head = reverse_LL(ll.head)

    return first, second


lst1 = LinkedList()
lst1.append(1)
lst1.append(2)
lst1.append(3)
lst1.append(4)
lst1.append(5)
lst1.append(6)
lst1.printList()
f1, s1 = sep(lst1)
f1.printList()
s1.printList()


lst2 = LinkedList()
lst2.append(1)
lst2.append(2)
lst2.append(3)
lst2.append(4)
lst2.append(5)
lst2.append(6)
lst2.append(7)
lst2.printList()
f2, s2 = sep(lst2)
f2.printList()
s2.printList()
```

**Q3**.	Write a program to create NÃ— N square matrix and each element is positive int type from 1 to N. In each row, the value of each element appeases exactly once and exactly once in each column as well. For example, if N = 3, the result will be as follows

1	2	3

2	3	1

3	1	2

	If N = 4, you will have 

1	2	3	4

2	3	4	1

3	4	1	2

4	1	2	3

 
 	Hint: rotate singly circular linked list in each row, or other method you want



```python
# Answer: 

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class CircularList:
    def __init__(self, data=None):
        self.head = None
        self.tail = None
        self.size = 0

    def clear(self):
        self.tail = None
        self.head = None

    def append(self, data):# we are adding data to the begining,so head is moving and tail is fixed point
        node = Node(data)
        if self.head: #if head is not none, for the first time it is none(we instantiate self.head=none, so we go to the else section)
            self.head.next = node #for the second time, we added one node so now self.head is not none and it means the previous node that we already instantiated)
            self.head = node
        else:
            self.head = node
            self.tail = node
        self.head.next = self.tail
        self.size += 1


def print_CL(nd):
    tmp = nd
    while 1:
        print(tmp.data, end=" ")
        tmp = tmp.next
        if tmp == nd:
            break


def createSquare(n):
    cl = CircularList()

    for i in range(0, n):
        cl.append(i+1)

    tmp = cl.tail
    for l in range(0, n):
        print_CL(tmp)
        tmp = tmp.next
        print(" ")


n = int(input("Enter Square size: "))
createSquare(n)
```

**Q4:** Given a doubly linked list with int type node, implement sorting operation by a function/method srt(ls), where ls is like

None<-Head<=>4<=>3<=>2<=>5<=>1->None. The new doubly linked list will be None<-Head<=>1<=>2<=>3<=> 4<=>5->None after the function/method call


```python
class Node(object):
    """ A Doubly-linked lists' node. """
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev


class DoublyLinkedList(object):
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append(self, data): 
        """ Append an item to the list. """

        new_node = Node(data, None, None)
        if self.head is None: 
            self.head = new_node
            self.tail = self.head
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

        self.count += 1

    def print_LL(self):
        current = self.head
        while current:
            print(str(current.data), end=" ")
            current = current.next
 
    def sortList(self):
        if self.head == None:
            return
        else:
            current = self.head
            while current.next is not None:
                index = current.next
                while index is not None:
                    if current.data > index.data:
                        temp = current.data
                        current.data = index.data
                        index.data = temp
                    index = index.next
                current = current.next


dll = DoublyLinkedList()
dll.append(4)
dll.append(3)
dll.append(2)
dll.append(5)
dll.append(1)
dll.print_LL()
dll.sortList()
print("\nAfter sorting")
dll.print_LL()
```
