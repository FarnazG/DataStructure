**Q1**.Write a program to insert an element to singly linked list in terms of reverse index, such as given
a=head->11->12->13-> NULL, after calling function/method insert_rev_index(a, 1, 123), and then new linked list will be head->11->12->123->13-> NULL. And if calling insert_rev_index(a, 0, 456), the new one should be head->11->12->123->13-> 456->NULL

```python

class Node:

    def __init__(self, data):
        self.data = data  # Assign data
        self.next = None  # Initialize next as null


# Linked List class contains a Node object
class LinkedList:

    def __init__(self):
        self.head = None

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def insertAfter(self, prev_node, new_data):
        if prev_node is None:
            print("The given previous node must inLinkedList.")
            return
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node

    def append(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return

        last = self.head
        while (last.next):
            last = last.next

        last.next = new_node

    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data, end=' ')
            temp = temp.next
        print("")

    def insert_rev_index(self, idx, value):
        tmp = Node(0)
        tmp.next = self.head

        h1 = tmp
        h2 = tmp
        new_node = Node(value)

        while idx > 0:
            h2 = h2.next
            idx -= 1
            if h2 is None:
                print("Index is beyond the list size")
                return

        while h2.next:
            h2 = h2.next
            h1 = h1.next

        new_node.next = h1.next
        h1.next = new_node
        self.head = tmp.next


lst1 = LinkedList()
lst1.append(11)
lst1.append(12)
lst1.append(13)
lst1.printList()
lst1.insert_rev_index(1, 123)
lst1.printList()
lst1.insert_rev_index(0, 456)
lst1.printList()
```

**Q2**.Create a function/method to concatenate two singly linked lists, like ls1= head->4->2->1-> NULL,
and ls2= head->5->7->3-> NULL.From function/method call concat_linked_list(ls1, ls2), the result is
head->4->2->1->5->7->3->NULL

```python
class Node:

    def __init__(self, data):
        self.data = data  # Assign data
        self.next = None  # Initialize next as null


# Linked List class contains a Node object
class LinkedList:

    def __init__(self):
        self.head = None

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def insertAfter(self, prev_node, new_data):
        if prev_node is None:
            print("The given previous node must inLinkedList.")
            return
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node

    def append(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return

        last = self.head
        while (last.next):
            last = last.next

        last.next = new_node

    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data, end=' ')
            temp = temp.next
        print("")


def concat_linked_list(lst1, lst2):
    tmp = Node(0)
    tmp.next = lst1.head
    h1 = tmp

    while h1.next:
        h1 = h1.next

    h1.next = lst2.head
    lst1.head = tmp.next
    return lst1


lst1 = LinkedList()
lst1.append(4)
lst1.append(2)
lst1.append(1)
lst1.printList()

lst2 = LinkedList()
lst2.append(5)
lst2.append(7)
lst2.append(3)
lst2.printList()

concat_linked_list(lst1, lst2).printList()
```

**Q3**.Find minimum value and maximum value by functions/methods for a given singly linked list with number or char type node
value, such as lst= head->3->1->2->NULL by calling a = min_num_linked_list(lst), a will be 1; if lst=head->C->A->B->NULL, and
a = min_str_linked_list(lst), a is A. Likewise, the same principle is for getting max value, like max_num_linked_list(lst) and
max_str_linked_list(lst)

```python
class Node:

    def __init__(self, data):
        self.data = data  # Assign data
        self.next = None  # Initialize next as null


# Linked List class contains a Node object
class LinkedList:

    def __init__(self):
        self.head = None

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def insertAfter(self, prev_node, new_data):
        if prev_node is None:
            print("The given previous node must inLinkedList.")
            return
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node

    def append(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return

        last = self.head
        while (last.next):
            last = last.next

        last.next = new_node

    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data, end=' ')
            temp = temp.next
        print("")


def min_num_linked_list(lst):
    h1 = lst.head
    min_value = 10000000
    while h1:
        if h1.data < min_value:
            min_value = h1.data
        h1 = h1.next

    return min_value


def max_num_linked_list(lst):
    h1 = lst.head
    max_value = -10000000
    while h1:
        if max_value < h1.data:
            max_value = h1.data
        h1 = h1.next

    return max_value


def min_str_linked_list(lst):
    h1 = lst.head
    min_value = 100000
    while h1:
        if ord(h1.data) < min_value:
            min_value = ord(h1.data)
        h1 = h1.next

    return chr(min_value)


def max_str_linked_list(lst):
    h1 = lst.head
    max_value = -10000000
    while h1:
        if max_value < ord(h1.data):
            max_value = ord(h1.data)
        h1 = h1.next

    return chr(max_value)


lst1 = LinkedList()
lst1.append(3)
lst1.append(1)
lst1.append(2)
print("Min char: " + str(min_num_linked_list(lst1)))
print("Max char: " + str(max_num_linked_list(lst1)))

lst2 = LinkedList()
lst2.append('C')
lst2.append('A')
lst2.append('B')
print("Min char: " + min_str_linked_list(lst2))
print("Max char: " + max_str_linked_list(lst2))
```

**Q4**.Write a program to swap positions of min and max node value in a given singly linked list. For instance, s1= head->3->1->2->NULL,
the new one will be head->1->3->2->NULL through swap_num_linked_list(s1); following the same rule for char node value, it is like
s2= head->C->A->B->NULL, calling swap_char_linked_list(s2) will get head->A ->C ->B->NULL

```python

class Node:

    def __init__(self, data):
        self.data = data  # Assign data
        self.next = None  # Initialize next as null


# Linked List class contains a Node object
class LinkedList:

    def __init__(self):
        self.head = None

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def insertAfter(self, prev_node, new_data):
        if prev_node is None:
            print("The given previous node must inLinkedList.")
            return
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node

    def append(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return

        last = self.head
        while (last.next):
            last = last.next

        last.next = new_node

    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data, end=' ')
            temp = temp.next
        print("")


def swap_num_linked_list(lst):
    h1 = lst.head
    min_node = Node(10000000)
    max_node = Node(-10000000)

    while h1:
        if h1.data < min_node.data:
            min_node = h1

        if max_node.data < h1.data:
            max_node = h1
        h1 = h1.next

    tmp = min_node.data
    min_node.data = max_node.data
    max_node.data = tmp
    return lst


def swap_str_linked_list(lst):
    h1 = lst.head
    min_node = Node(chr(1114111))
    max_node = Node(chr(0))

    while h1:
        if ord(h1.data) < ord(min_node.data):
            min_node = h1

        if ord(max_node.data) < ord(h1.data):
            max_node = h1
        h1 = h1.next

    tmp = min_node.data
    min_node.data = max_node.data
    max_node.data = tmp
    return lst


lst1 = LinkedList()
lst1.append(3)
lst1.append(1)
lst1.append(2)
lst1.printList()
swap_num_linked_list(lst1).printList()

lst2 = LinkedList()
lst2.append('C')
lst2.append('A')
lst2.append('B')
lst2.printList()
swap_str_linked_list(lst2).printList()
```
